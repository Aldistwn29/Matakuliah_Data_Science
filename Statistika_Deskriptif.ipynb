```python
import statistics as stats
import pandas as pd
import numpy as np
```

# **Tugas 1**

```python
data_1 = {
          'Modal':[(10, 10.9), (11, 11.9), (12, 12.9), (13, 13.9), (14, 14.9), (15, 15.9)],
          'Banyak_Biro_Perjalan_perunit':[8, 12, 15, 32, 48, 10]
          }
df_1 = pd.DataFrame(data_1)
```

```python
df_1.head()
```

```python
# A. Tentukan Mean, Modus, Dan Median
mean = np.mean(df_1['Banyak_Biro_Perjalan_perunit'])
median = np.median(df_1['Banyak_Biro_Perjalan_perunit'])
modus = stats.mode(df_1['Banyak_Biro_Perjalan_perunit'])
print(f"Mean : {mean}")
print(f"Median : {median}")
print(f"Modus : {modus}\n")

# Secara Manual
nilai = df_1['Banyak_Biro_Perjalan_perunit']
# Mean
mean_manual = sum(nilai)/len(nilai)
print(f'mean_manual:{mean_manual}')
# Median
data_urut = sorted(nilai)
n = len(data_urut)
if n % 2 ==1:
    median_manual = data_urut(n//2)
else:
    median_manual = (data_urut[n//2-1]+data_urut[n//2])//2
print(f'median_manual:{median_manual}')
# Modus
modus_manual = nilai.value_counts()
print(f'modus_manual:{modus_manual}')
```

```python
import matplotlib.pyplot as plt

# Menghitung rata-rata modal
df_1['Rata-rata_modal'] = df_1['Modal'].apply(lambda x: np.mean(x))
# B. Tujunakan Mean, Median dan Modus dalam histogram
plt.figure(figsize=(10, 6))
plt.bar(df_1['Rata-rata_modal'], df_1['Banyak_Biro_Perjalan_perunit'], color='blue', alpha=0.7)

# Menambhakan garis
plt.axhline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean:{mean:.2f}')
plt.axhline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median:{median:.2f}')
plt.axhline(modus, color='black', linestyle='dashed', linewidth=2, label=f'Modus:{modus}')

plt.title('Histogram Banyak Biro Perjalan per Unit')
plt.xlabel('modal')
plt.ylabel('Banyak Biro Perjalan per Unit')
plt.legend()
plt.grid(True)
plt.show()
```

# **Tugas 3**

```python
data_3 = {
          'Nilai_Ebtanas':[(20, 29), (30, 39), (40, 49), (50, 59), (60, 69), (70, 79), (80, 89), (90, 99)],
          'Banyak_Siswa':[17, 12, 10, 38, 66, 40, 12, 5]
          }
df_3 = pd.DataFrame(data_3)
```

```python
df_3.head()
```

```python
# soal a
# menghitung jumlah siswa
total_siswa = df_3['Banyak_Siswa'].sum()
print(f'Jumlah Siswa: {total_siswa}')

# menghitung siswa yang mendapatkan beasiswa 5 persen
persentase_beasiswa = int(total_siswa * 0.05)
print(f'Jumlah Siswa yang mendapatkan Beasiswa: {jumlah_beasiswa}')

# membuat daftar rata-rata nilai untuk beasiswa
nilai_siswa = []
for index, row in df_3.iterrows():
    batas_atas = row['Nilai_Ebtanas'][1]
    for _ in range(row['Banyak_Siswa']):
      nilai_siswa.append(batas_atas)
# Mengurut kan nilai yg tertinggi - terkecil
nilai_siswa_tertinggi = sorted(nilai_siswa, reverse=True)

# Dapatkan nilai terendah dari siswa dengan beasiswa
nilai_terendah_beasiswa = nilai_siswa_tertinggi[persentase_beasiswa - 1] if persentase_beasiswa > 0 else None
print(f'Nilai terendah dari siswa dengan beasiswa: {nilai_terendah_beasiswa}')
```

```python
# Soal b
persentase_lulus = int(total_siswa * 0.6)
nilai_terendah_lulus = nilai_siswa_tertinggi[persentase_lulus - 1] if persentase_lulus > 0 else None
print(f"jumlah_siswa_lulus: {persentase_lulus}")
print(f"nilai_terendah_lulus : {nilai_terendah_lulus }")
```

```python
# Soal c
nilai_siswa_terendah = sorted(nilai_siswa)
persentasi_tidak_lulus = int(total_siswa * 0.2)
nilai_tertinggi_tidak_lulus = nilai_siswa_terendah[persentasi_tidak_lulus - 1] if persentasi_tidak_lulus > 0 else None
print(f"jumlah siswa tidak lulus: {persentasi_tidak_lulus}")
print(f"Nilai tertinggi tidak lulus : {nilai_tertinggi_tidak_lulus }")
```

# **Tugas 4**

Hasil survey terhadap 120 layanan hotel di suatu daerah

```python
data_4 = {
          'Nilai_Layanan':[(40,49), (50,59), (60,69), (70,79), (80,89), (90,99)],
          'Banyak_Hotel(unit)':[5, 25, 15, 50, 20, 5]
          }
df_4 = pd.DataFrame(data_4)
```

```python
df_4
```

```python
import matplotlib.pyplot as plt
```

```python
df_4['Rata-rata_layanan'] = df_4['Nilai_Layanan'].apply(lambda x: np.mean(x))
plt.bar(df_4['Rata-rata_layanan'], df_4['Banyak_Hotel(unit)'])
plt.xlabel('Nilai Layanan')
plt.ylabel('Banyak Hotel(unit)')
plt.title('Distribusi nilai layanan hotel')
plt.show()
```

```python
# Soal a
# Menghitung jumlah siswa
jumlah_total_hotel = df_4['Banyak_Hotel(unit)'].sum()

# Menghitung 20 % dari total keseluruhan
prcent_20 = int(jumlah_total_hotel * 0.2)

# mengurutkan nilai layanan dari tertinggi ke terendah
df_4 = df_4.sort_values(by='Nilai_Layanan', ascending=False).reset_index(drop=True)

# Mengambil nilai yg terendah
commulative_hotel = 0
threshold_value = None

for index, row in df_4.iterrows():
  commulative_hotel += row['Banyak_Hotel(unit)']
  if commulative_hotel >= prcent_20:
    threshold_value = row['Nilai_Layanan'][0]
    break

print("Total seluruh siswa: ", jumlah_total_hotel)
print("20% dari total siswa: ", prcent_20)
print("Nilai Terendah untuk penghargaan: ", threshold_value)
```

```python
# Soal b
# Mencari berpa banyak hotel dari 30%
bottom_30 = int(jumlah_total_hotel * 0.3)

# menggurutkan nilai layanan dari yg terendah ke terbesar
df_4 = df_4.sort_values(by='Nilai_Layanan').reset_index(drop=True)

# mencari nilai tertinggi
commulative_hotel_2 = 0
threshold_value_2 = None

for index, row in df_4.iterrows():
  commulative_hotel_2 += row['Banyak_Hotel(unit)']
  if commulative_hotel_2 >= bottom_30:
    threshold_value_2 = row['Nilai_Layanan'][1]
    break

print("Jumlah Hotel :",jumlah_total_hotel)
print("30% dari total hotel :",bottom_30)
print("Nilai tertinggi yg akan diberikan bantuan :",threshold_value_2)
```
